```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("heart_disease.csv")

#basic insights
rows,columns=data.shape
print(f"rows:{rows}")
print(f"columns:{columns}")
print()

print(data.describe())
uniqueCountG=data['Gender'].nunique()
uniqueCountS=data['Smoking'].nunique()
print(f"Unique values in Gender Column:{uniqueCountG}")
print(f"Unique values in Smoking Column:{uniqueCountS}")
print()
print()

#Handling Missing data
print("Handling Missing Values:")
print()
missingCount=data.isnull().sum()
print(missingCount)
#Method to fill the empty places
# - Mean: Used for numerical data when values are missing at random.  
# - Median: Used when data has outliers to avoid skewing.  
# - Mode: Used for categorical data to replace missing values with the most frequent value.  

#Data visualization
print()
print("Histogram for Age:")
plt.figure(figsize=(4,4))
sns.histplot(data['Age'])
plt.title('Age Distribution of Heart Disease Patients')
plt.xlabel('Age(Years)')
plt.ylabel('Number of people')
plt.show()

heart_disease_counts = data['Heart Disease Status'].value_counts()

# Map 0 to 'No' and 1 to 'Yes'
labels = ["No", "Yes"]

#bar chart
plt.figure(figsize=(6, 4))
sns.barplot(x=labels, y=heart_disease_counts.values, palette=['blue', 'red'])
plt.xlabel("Heart Disease Status")
plt.ylabel("Number of People")
plt.title("Count of People With and Without Heart Disease")

plt.show()
print()

#boxplot for cholestrol levels
print("box-plot(Cholesterol Level):")
plt.figure(figsize=(6, 5))
sns.boxplot(x=data['Cholesterol Level'],vert=False, patch_artist=True, 
            boxprops=dict(facecolor="cyan"))
plt.xlabel('Cholesterol Level')
plt.ylabel('Number Of People')
plt.title("box plot of Cholesterol Levels")
plt.show()
```
